```python
import pandas as pd
import json

json_path = "export.json"

with open(json_path, "r") as f:
    data_json = json.load(f)

df_json = pd.DataFrame(data_json)

print('---------------------<info()>----------------------')
print(df_json.info())
print('---------------------<sum()>----------------------')
print(df_json.isnull().sum())
print('---------------------<dtypes>----------------------')
print(df_json.dtypes)
print('---------------------<describe()>----------------------')
print(df_json.describe())
print('---------------------<>----------------------')
print(df_json.info)
```

```python
import pandas as pd
import xml.etree.ElementTree as ET

with open("export.xml", "r", encoding="utf-8") as f:
    xml_content = f.read()

with open("fixed_export.xml", "w", encoding="utf-8") as f:
    f.write(xml_content)


xml_path = "fixed_export.xml"

tree = ET.parse(xml_path)
root = tree.getroot()

elements = root.findall('ROW')

df = pd.DataFrame(columns=['tier', 'veteran', 'wins', 'rank', 'summonerid', 'losses'])

for element in elements:
    row = {
      'tier' : element.find('tier').text,
      'veteran' : element.find('veteran').text,
      'wins' : element.find('wins').text,
      'rank' : element.find('rank').text,
      'summonerid' : element.find('summonerid').text,
      'losses' : element.find('losses').text,
    }
    new_df = pd.DataFrame([row], columns=df.columns)
    df_xml = pd.concat([df, new_df], ignore_index=True)

print('---------------------<info()>----------------------')
print(df_xml.info())
print('---------------------<sum()>----------------------')
print(df_xml.isnull().sum())
print('---------------------<dtypes>----------------------')
print(df_xml.dtypes)
print('---------------------<describe()>----------------------')
print(df_xml.describe())
print('---------------------<>----------------------')
print(df_xml.info)
```

```python
import pandas as pd
from lxml import etree

xml_path = "fixed_export.xml"

# lxml을 사용하여 XML 파싱
tree = etree.parse(xml_path)
root = tree.getroot()

elements = root.findall('ROW')

# 데이터를 저장할 리스트
rows = []

for element in elements:
    row = {
        'tier': element.find('tier').text if element.find('tier') is not None else None,
        'veteran': element.find('veteran').text if element.find('veteran') is not None else None,
        'wins': element.find('wins').text if element.find('wins') is not None else None,
        'rank': element.find('rank').text if element.find('rank') is not None else None,
        'summonerid': element.find('summonerid').text if element.find('summonerid') is not None else None,
        'losses': element.find('losses').text if element.find('losses') is not None else None,
    }
    rows.append(row)

# 한 번에 DataFrame 생성
df = pd.DataFrame(rows, columns=['tier', 'veteran', 'wins', 'rank', 'summonerid', 'losses'])

print('---------------------<info()>----------------------')
print(df.info())
print('---------------------<sum()>----------------------')
print(df.isnull().sum())
print('---------------------<dtypes>----------------------')
print(df.dtypes)
print('---------------------<describe()>----------------------')
print(df.describe())
print('---------------------<>----------------------')
```

```python
import pandas as pd
from lxml import etree

xml_path = "fixed_export.xml"

# lxml을 사용하여 XML 파싱
tree = etree.parse(xml_path)
root = tree.getroot()

elements = root.findall('ROW')

df = pd.DataFrame(columns=['tier', 'veteran', 'wins', 'rank', 'summonerid', 'losses'])

for element in elements:
    # 각 항목을 찾을 때 None 처리 추가
    row = {
        'tier': element.find('tier').text if element.find('tier') is not None else None,
        'veteran': element.find('veteran').text if element.find('veteran') is not None else None,
        'wins': element.find('wins').text if element.find('wins') is not None else None,
        'rank': element.find('rank').text if element.find('rank') is not None else None,
        'summonerid': element.find('summonerid').text if element.find('summonerid') is not None else None,
        'losses': element.find('losses').text if element.find('losses') is not None else None,
    }
    new_df = pd.DataFrame([row], columns=df.columns)
    df = pd.concat([df, new_df], ignore_index=True)

print('---------------------<info()>----------------------')
print(df.info())
print('---------------------<sum()>----------------------')
print(df.isnull().sum())
print('---------------------<dtypes>----------------------')
print(df.dtypes)
print('---------------------<describe()>----------------------')
print(df.describe())
print('---------------------<>----------------------')
```
