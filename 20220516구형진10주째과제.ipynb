```python
# 1. 패키지 로딩
import pandas as pd
import matplotlib.pyplot as plt

# 2. 데이터 로딩
path = '/owid-covid-data.csv'
df = pd.read_csv(path)

# 3. 국가별 필터링
kor_df = df[df['location'] == 'South Korea'].set_index('date')
fra_df = df[df['location'] == 'France'].set_index('date')

# 4. 확진자 및 인구 데이터
kor_cases = kor_df['total_cases']
fra_cases = fra_df['total_cases']
kor_pop = kor_df['population']
fra_pop = fra_df['population']

# 5. 인구 대비 비율 계산
kor_ratio = kor_cases / kor_pop
fra_ratio = fra_cases / fra_pop

# 6. 비교 데이터프레임 생성
compare_df = pd.DataFrame({
    'KOR': kor_ratio,
    'FRA': fra_ratio
})

# 7. 전체 기간 시각화
compare_df.plot(title='COVID-19 Infection Ratio: South Korea vs France', figsize=(12, 6))
plt.xlabel('Date')
plt.ylabel('Confirmed Cases per Person')
plt.xticks(rotation=45)
plt.grid()
plt.show()

# 8. 2022년만 시각화
compare_df['2022-01-01':'2022-12-31'].plot(title='2022 Infection Ratio: South Korea vs France', figsize=(12, 6))
plt.xlabel('Date')
plt.ylabel('Confirmed Cases per Person')
plt.xticks(rotation=45)
plt.grid()
plt.show()
```
